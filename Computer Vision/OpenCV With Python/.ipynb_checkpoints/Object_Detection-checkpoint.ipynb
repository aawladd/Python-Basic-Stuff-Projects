# Template Matching

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
full = cv2.imread('Computer-Vision-with-Python/DATA/sammy.jpg')
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
face = cv2.imread('Computer-Vision-with-Python/DATA/sammy_face.jpg')
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(face)
```

```python
full.shape
```

```python
face.shape
```

```python
mystring = 'sum'
```

```python
myfunc = eval(mystring)
```

```python
myfunc([2,3,4])
```

```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    
    #Create a copy
    full_copy = full.copy()
    
    method = eval(m)
    
    # Template Matching
    res = cv2.matchTemplate(full_copy, face, method)
    
    min_val , max_val , min_loc, max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
        
    height, width , channels = face.shape
    
    bottom_right = (top_left[0]+width, top_left[1] + height)
    
    cv2.rectangle(full_copy, top_left, bottom_right, (255,0,0), 10)
    
    #PLT AND SHOW
    plt.subplot(121)
    plt.imshow(res)
    plt.title('HEAT MAP OF TEMPLATE MATCHING')
    
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('DETECTION OF TEMPLATE')
    #Title with the Method Used
    plt.suptitle(m)
    
    plt.show()
    
    print('\n\n')
```

```python
my_method = eval('cv2.TM_CCOEFF')
res = cv2.matchTemplate(full, face, my_method)
```

```python
plt.imshow(res)
```

# Corner Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('Computer-Vision-with-Python/DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat_chess, cmap = 'gray')
```

```python
real_chess = cv2.imread('Computer-Vision-with-Python/DATA/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_real_chess, cmap = 'gray')
```

```python
gray= np.float32(gray_flat_chess)
```

```python
dst = cv2.cornerHarris(src = gray, blockSize = 2, ksize = 3, k =0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
flat_chess[dst>0.01*dst.max()] = [255,0,0] #RGB
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)
dst = cv2.cornerHarris(src = gray, blockSize = 2, ksize = 3, k= 0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
real_chess[dst>0.01*dst.max()] = [255,0,0]
```

```python
plt.subplot(111)
plt.imshow(real_chess)
plt.show()
```

```python
edges = cv2.Canny(image = img, threshold1 = 0, threshold2 = 255)
```

```python
plt.imshow(edges)
```

```python
med_value = np.median(img)
```

```python
med_value
```

```python
#LOWER THRESHOLD TO EITHER 0 OR 705 OF THE MEDIAN VALUE WHICH EVER IS GREATER
lower = int(max(0, 0.7*med_value))
upper = int(min(255, 1.3*med_value))
```

```python
edges = cv2.Canny(image = img, threshold1 = lower, threshold2 = upper+100)
plt.imshow(edges)
```

```python
upper
```

```python
lower
```

```python
blurred_img = cv2.blur(img, ksize = (5,5))
plt.imshow(blurred_img)
```

```python
edges = cv2.Canny(image = blurred_img, threshold1 = lower, threshold2 = upper+50)
plt.imshow(edges)
```

```python
cv2.imwrite('Dog_Blurred_edges.jpg', edges)
```

# Feature Mathching

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap = 'gray'):
    fig = plt.figure(figsize = (12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
reeses = cv2.imread('Computer-Vision-with-Python/DATA/reeses_puffs.png', 0)
display(reeses)
```

```python
cereals = cv2.imread('Computer-Vision-with-Python/DATA/many_cereals.jpg', 0)
display(cereals)
```

```python
orb = cv2.ORB_create()
```

```python
kp1, des1 =  orb.detectAndCompute(reeses, None)
kp2, des2 =  orb.detectAndCompute(cereals, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)
```

```python
matches = bf.match(des1, des2)
```

```python
single_match = matches[0]
```

```python
single_match.distance
```

```python
matches = sorted(matches, key=lambda x: x.distance)
```

```python
matches[0].distance
```

```python
reses_matches = cv2.drawMatches(reeses, kp1, cereals, kp2, matches[:25], None, flags = 2)
```

```python
display(reses_matches)
```
